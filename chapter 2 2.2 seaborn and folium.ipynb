## Regression plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

dataset=pd.read_csv('auto-mpg.csv')
sns.regplot(x='mpg',y='acceleration',data=dataset)
plt.show()
```

```python
dataset.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

dataset=pd.read_csv('auto-mpg.csv')
sns.regplot(x='cylinders',y='displacement',data=dataset)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

dataset=pd.read_csv('auto-mpg.csv')
sns.regplot(x='model year',y='origin',data=dataset)
plt.show()
dataset.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

dataset=pd.read_csv('auto-mpg.csv')
sns.set(font_scale=1.5)
sns.regplot(x='mpg',y='acceleration',data=dataset,color='green',marker='^',scatter_kws={'s':200})
plt.xlabel('mpg')
plt.ylabel('acceleration')
plt.title('mpg vs acceleration')
plt.show()
```

## Heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

data=np.random.randint(1,100,size=(10,10))
print(data)
sns.heatmap(data)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

data=np.random.randint(1,100,size=(10,10))
print(data)
sns.heatmap(data,vmin=30,vmax=70)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

data=np.random.randint(1,100,size=(10,10))
print(data)
sns.heatmap(data,cmap='tab20')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

data=np.random.randint(1,100,size=(10,10))
print(data)
sns.heatmap(data,cmap='tab20',annot=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

data=np.random.randint(1,100,size=(10,10))
print(data)
sns.heatmap(data,cmap='tab20',linewidths=2,linecolor='black',cbar=False,xticklabels=False,yticklabels=False)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

df=pd.DataFrame(np.random.random((5,5)),columns=['a','b','c','d','e'])
print(df)
sns.heatmap(df)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

plt.figure(figsize=(10,3))
data=np.random.randint(1,100,size=(3,10))
print(data)
sns.heatmap(data,cmap='tab20',annot=True)
plt.show()
```
