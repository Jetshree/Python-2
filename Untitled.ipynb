```python
import requests,json
def get_coordinates():
    api_key = input("enter api key")
    city = input("enter city name")
    request_url = f"http://api.openweathermap.org/geo/1.0/direct?q={city}&appid={api_key}"
    response = requests.get(request_url)
    coordinates = response.json()
    print(coordinates)
    lattitude = coordinates[0]['lat']
    longitude = coordinates[0]['lon']
    return (lattitude,longitude)
    
lat,lon = get_coordinates()
print(lat,lon)
```

```python
def get_current_weather():
    api_key = input("enter api key")
    latitude = lat
    longitude = lon
    request_url = f"https://api.openweathermap.org/data/2.5/weather?lat={latitude}&lon={longitude}&appid={api_key}"
    response = requests.get(request_url)
    weather = response.json()
    return weather
weather = get_current_weather()
print(json.dumps(weather,indent =5))
```

```python
def air_pollution():
    api_key = input("enter api key")
    latitude = lat
    longitude = lon
    request_url = f"http://api.openweathermap.org/data/2.5/air_pollution?lat={latitude}&lon={longitude}&appid={api_key}"
    response = requests.get(request_url)
    air_pollution = response.json() 
    print(air_pollution)
    if(air_pollution["list"][0]["main"]["aqi"] == 1):
        return "air quality is good"
    
    elif(air_pollution["list"][0]["main"]["aqi"] == 2):
        return "air quality is Fair"
    
    elif(air_pollution["list"][0]["main"]["aqi"]== 3):
        return "air quality is Moderate"
    
    elif(air_pollution["list"][0]["main"]["aqi"] == 4):
        return "air quality is Poor"
    
    elif(air_pollution["list"][0]["main"]["aqi"]== 5):
        return "air quality is Very Poor"
    

text = air_pollution()
print(text)
```

```python
request_url =f"http://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={api_key}"
response = requests.get(request_url)
air_pollution = response.json()
print(json.dumps(air_pollution,indent =5))
```

```python
date_time =[]
temp =[]
pressure=[]
humidity=[]
weather_desc=[]
for i in air_pollution["list"]:
    date_time.append(i["dt_txt"])
    temp.append(i["main"]["temp"])
    pressure.append(i["main"]["pressure"])
    humidity.append(i["main"]["humidity"])
    weather_desc.append(i["weather"][0]["description"])

    
data={"date_time":date_time,"temp":temp,"pressure":pressure,"humidity":humidity,"weather_desc":weather_desc}
import pandas as pd
dataframe = pd.DataFrame(data)
dataframe
```
