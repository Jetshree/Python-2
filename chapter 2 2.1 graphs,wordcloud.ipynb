```python
import matplotlib.pyplot as plt
y=[1,5,8,9,2,0,3,10,4,7]
x=range(1,11)
plt.annotate(text='First Entry',xy=[1,1])
plt.annotate(text='Last Entry',xy=[10,7])
plt.plot(x,y)
plt.xlabel("Entries")
plt.ylabel('Values')
plt.title('entry vs values')
plt.show()
```

```python
import matplotlib.pyplot as plt
v1=[1,3,6,8,10,4,7]
v2=[5,8,1,9,3,10,7]
plt.plot(range(1,8),v1,label='First')
plt.plot(range(1,8),v2,label='second')
plt.legend(loc=0)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=range(1,6)
y=[1,4,6,8,4]
plt.fill_between(x,y,color='blue',alpha=0.4)
plt.plot(x,y,color='red',alpha=0.8)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

time=np.arange(12)
income=np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expense=np.array([6,6,8,3,6,9,7,8,6,6,4,8])
plt.plot(time,income,color='green')
plt.plot(time,expense,color='red')
plt.fill_between(time,income,expense,where=(income>expense),color='green',alpha=0.25,label='positive')
plt.fill_between(time,income,expense,where=(income<=expense),color='red',alpha=0.25,label='negative')
plt.xlabel('time')
plt.ylabel('income/expense')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

time=np.arange(12)
income=np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expense=np.array([6,6,8,3,6,9,7,8,6,6,4,8])
plt.plot(time,income,color='green')
plt.plot(time,expense,color='red')
plt.fill_between(time,income,expense,where=(income>expense),color='green',alpha=0.25,label='positive',interpolate=True)
plt.fill_between(time,income,expense,where=(income<=expense),color='red',alpha=0.25,label='negative',interpolate=True)
plt.xlabel('time')
plt.ylabel('income/expense')
plt.legend()
plt.show()
```

### Stacked area chart

```python
x=range(1,6)
y1=[1,4,6,8,9]
y2=[2,2,7,10,12]
y3=[2,8,5,10,6]
plt.stackplot(x,y1,y2,y3,labels=['a','b','c','d'],colors=['yellow','black','blue'])
plt.legend(loc='upper left')
plt.show()
```

```python
x=range(1,6)
y1=[2,2,2,2,2]
y2=[2,2,2,2,2]
y3=[2,2,2,2,2]
plt.stackplot(x,y1,y2,y3,labels=['a','b','c','d'],colors=['lavender','skyblue','pink'])
plt.legend(loc='upper left')
plt.show()
```

```python
from wordcloud import WordCloud,STOPWORDS
import matplotlib.pyplot as plt
alice=open('alice_in_wonderland.txt')
alice=alice.read()
stop=set(STOPWORDS)
print(stop)
print(len(stop))
stop.add('said')
```

```python
alice_wc=WordCloud(background_color='white',max_words=2000,stopwords=stop,collocations=False)
alice_wc.generate(alice)
plt.imshow(alice_wc)
plt.axis('off')
plt.show()
```
