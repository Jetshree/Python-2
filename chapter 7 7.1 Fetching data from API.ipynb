```python
# http://api.openweathermap.org/data/2.5/air_pollution?lat={lat}&lon={lon}&appid={API key}
    
import requests, json

api_key = input("Enter api key: ")

def get_coordinates():
    city = input("Enter city name: ")
    request_url = f"http://api.openweathermap.org/geo/1.0/direct?q={city}&appid={api_key}"
    response = requests.get(request_url)
    coordinates=response.json()
    lattitude = coordinates[0]['lat']
    longitude = coordinates[0]['lon']
    return (lattitude, longitude)
lat,lon = get_coordinates()

def get_aqi(lat,lon):
    request_url = f"http://api.openweathermap.org/data/2.5/air_pollution?lat={lat}&lon={lon}&appid={api_key}"
    response = requests.get(request_url)
    aq_index = response.json()
    if aq_index['list'][0]['main']['aqi']==1:
        print("AQI: Good")
    elif aq_index['list'][0]['main']['aqi']==2:
        print("AQI: Fair")
    elif aq_index['list'][0]['main']['aqi']==3:
        print("AQI: Moderate")
    elif aq_index['list'][0]['main']['aqi']==4:
        print("AQI: Poor")
    elif aq_index['list'][0]['main']['aqi']==5:
        print("AQI: Very poor")
    
get_aqi(lat,lon)
```
