```python
import pandas as pd
s=pd.read_csv("US_Unemployment_Oct2012.csv")
print(s.head())

state_geo="us-states.json"
```

```python
import folium
usa_state=folium.Map(location=[48,-122],zoom_start=3)
folium.Choropleth(geo_data=state_geo,name='choropleth',data=s,
                  columns=['State','Unemployment'],key_on='feature.id',fill_color='YlGnBu',fill_opacity=0.7,
                  line_opacity=0.3,legend_name='Unemployment Scale').add_to(usa_state)

usa_state
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G=nx.Graph()
G.add_node(1)
G.add_nodes_from([2,3,4,5,6])
G.add_edge(1,2)
G.add_edge(1,1)
G.add_edges_from([(2,3),(3,6),(4,6),(5,6)])
nx.draw_networkx(G,node_size=850,node_color='orange',width=2,edge_color='blue')
```

```python
G=nx.DiGraph()
l=[[0,1,0,1],[0,0,0,1],[1,0,0,0],[1,0,1,0]]
for i in range(4):
    for j in range(4):
        if l[i][j]==1:
            G.add_edge(i,j)
nx.draw_networkx(G,node_size=850,node_color='red',width=2,edge_color='blue')
```

```python
G=nx.DiGraph()
l=[[1,0,0,1,0,1,1],[0,1,1,1,1,1,0],[1,1,1,0,1,0,1]]
G.add_nodes_from([0,1,2])
for i in range(3):
    for j in range(7):
        if l[i][j]==1:
            G.add_edge(i,j)
nx.draw_networkx(G,node_size=850,node_color='red',width=2,edge_color='blue')
```
