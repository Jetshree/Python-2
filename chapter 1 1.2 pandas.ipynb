## Dataframe

- 2D structure like 2D array or a table with rows and columns

```python
import pandas as pd

data={"calories": [400,300,390],"duration": [50,40,45]}
df=pd.DataFrame(data)
print(df)
print(df.get("calories"))
```

```python
print(df.loc[0])
print(df.iloc[0])
```

```python
df.loc[[0,1]]
```

```python
import pandas as pd

df=pd.DataFrame(data)
print(df)
print(df.loc[0])
print(df.loc[0:2]) # includes the end point
print(df.iloc[0:2]) # excludes the end point
```

## Parallel Coordinates

```python
import pandas as pd
a=pd.read_csv("auto-mpg.csv")
```

```python
from pandas.plotting import parallel_coordinates
pll=parallel_coordinates(a,'cylinders',cols=['acceleration','mpg'],color=['red','blue','green','pink'])
```

```python
from pandas.plotting import parallel_coordinates
pll=parallel_coordinates(a,'displacement',cols=['acceleration','mpg'])
```

```python
from pandas.plotting import parallel_coordinates
pll=parallel_coordinates(a,'cylinders',cols=['cylinders','mpg','origin'],color=['red','blue','green','pink'])
```
